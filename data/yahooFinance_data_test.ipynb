```python
import yfinance as yf
from datetime import datetime
import mplfinance as mpf
```

```python
# Define the ticker symbol and date range
ticker_symbol = "RELIANCE.NS"
start_date = "2025-01-01"
end_date = datetime.now().strftime('%Y-%m-%d')  # Current date

# Fetch historical data
data = yf.download(ticker_symbol, start=start_date, end=end_date, interval="1d")

# Display the data
print(data)
```

```python
# Prepare the data for mplfinance
data.index.name = 'Date'  # Ensure the index is named 'Date'
data.columns = data.columns.droplevel(1)  # Drop the multi-level column index
```

```python
# Plot the candlestick chart
mpf.plot(data, type='candle', style='charles', volume=False, mav = (10,20,50,200), figsize=(14, 8), title=f"Candlestick Chart for {ticker_symbol}")
```
